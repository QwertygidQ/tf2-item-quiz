```python
import requests
from json import dumps
```

```python
def get_weapon_info(weapon):
    r = requests.get('https://wiki.teamfortress.com/w/api.php?action=query&prop=revisions&rvprop=content&format=json&titles=' + weapon + '&rvsection=0&redirects')

    json = r.json()
    json = json['query']['pages']
    json = json[list(json.keys())[0]]
    title = json['title']
    json = json['revisions'][0]['*']
    info = json.split('\n')

    info = list(map(str.strip, info))
    info = list(filter(lambda x: len(x) > 0 and x[0] == '|', info))
    info = list(map(lambda x: x[2:], info))

    d = {}
    for obj in info:
        temp = list(map(str.strip, obj.split('=')))
        d[temp[0]] = temp[1]
    d['name'] = title

    proper_dict = {
        'name': d['name'],
        'tfclass': d['used-by'][2:-2],
        'level': 'Level ' + d['level'] + ' ' + d['item-kind']
    }
    attrs = list(filter(lambda k: k.startswith('att'), d))
    proper_dict['attrs'] = list(map(lambda x: {'type': x.split('-')[-1], 'text': d[x]}, attrs))

    r = requests.get('https://wiki.teamfortress.com/w/api.php?action=query&prop=images&format=json&titles=' + title)

    json = r.json()
    json = json['query']['pages']
    json = json[list(json.keys())[0]]['images']

    image = next(x for x in json if x['title'].startswith('File:Backpack'))['title']
    
    r = requests.get('https://wiki.teamfortress.com/w/api.php?action=query&titles=' + image + '&prop=imageinfo&iiprop=url&format=json')
    json = r.json()
    json = json['query']['pages']
    image = json[list(json.keys())[0]]['imageinfo'][0]['url']
    
    proper_dict['img'] = image
    return proper_dict
```

```python
r = requests.get('https://wiki.teamfortress.com/w/api.php?action=query&titles=Weapons&prop=links&format=json&redirects')
tmp_json = r.json()
tmp_json = tmp_json['query']['pages']
tmp_json = tmp_json[list(tmp_json.keys())[0]]['links']
weapons = [*weapons, *tmp_json]

plcontinue = '1585|0|Apoco-Fists'
while plcontinue != '':
    r = requests.get('https://wiki.teamfortress.com/w/api.php?action=query&titles=Weapons&prop=links&format=json&redirects&plcontinue=' + plcontinue)
    tmp_json = r.json()
    try:
        plcontinue = tmp_json['continue']['plcontinue']
    except:
        break
    tmp_json = tmp_json['query']['pages']
    tmp_json = tmp_json[list(tmp_json.keys())[0]]['links']
    print(plcontinue)
    weapons = [*weapons, *tmp_json]
```

```python
links = list(map(lambda x: x['title'], weapons))
links
```
